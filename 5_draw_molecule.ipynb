```python
from rdkit import Chem
from rdkit.Chem import Draw
import pandas as pd
import numpy as np
```

```python
Borane_Radical_df = pd.read_csv(r'Data\csvs\reactants_B.csv')
molecules = [Chem.MolFromSmiles(smiles) for smiles in Borane_Radical_df['Smiles']]
Indexs = [f"{index:05}" for index in Borane_Radical_df.Index.to_numpy()]
print(len(molecules))
```

```python
for start_id in range(0, len(molecules), 48):
    end_id = start_id + 48 if start_id + 48 < len(molecules) else len(molecules)
    img = Draw.MolsToGridImage(molecules[start_id:end_id], legends=Indexs[start_id:end_id], molsPerRow=6, subImgSize=(200, 200), useSVG=True)
    with open(f"Figure\\FigureInSI\\Borane_Radical_{start_id:02}to{end_id:02}.svg", "w") as f:
        f.write(img)
```
