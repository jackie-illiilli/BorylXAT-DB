```python
from rdkit import Chem
from rdkit.Chem import Draw
import pandas as pd
import numpy as np
from DFTStructureGenerator.project_paths import CSV_DIR, FIGURE_DIR

FIGURE_IN_SI_DIR = FIGURE_DIR / "FigureInSI"
FIGURE_IN_SI_DIR.mkdir(parents=True, exist_ok=True)
```

```python
Borane_Radical_df = pd.read_csv(CSV_DIR / "reactants_B.csv")
molecules = [Chem.MolFromSmiles(smiles) for smiles in Borane_Radical_df['Smiles']]
Indexs = [f"{index:05}" for index in Borane_Radical_df.Index.to_numpy()]
print(len(molecules))
```

```python
for start_id in range(0, len(molecules), 48):
    end_id = start_id + 48 if start_id + 48 < len(molecules) else len(molecules)
    img = Draw.MolsToGridImage(molecules[start_id:end_id], legends=Indexs[start_id:end_id], molsPerRow=6, subImgSize=(200, 200), useSVG=True)
    with open(FIGURE_IN_SI_DIR / f"Borane_Radical_{start_id:02}to{end_id:02}.svg", "w") as f:
        f.write(img)
```

```python
from DFTStructureGenerator.draw import plot_scatter_fit, plot_distribution, plot_panel
from matplotlib import pyplot as plt
fig = plt.figure(figsize=(5,3), dpi=300)
ax = fig.add_subplot()
color_main = "#548235"
color_fill = "#e2f0d9"
color_2_main = "#c04f15"
color_2_fill = "#f2aa84"
Cl_df = pd.read_csv(CSV_DIR / "reactants_Cl.csv")
sp2_C = []
sp3_C = []
for row_id, row in Cl_df.iterrows():
    if Chem.MolFromSmiles(row['Smiles']).HasSubstructMatch(Chem.MolFromSmarts("cCl")):
        sp2_C.append(row_id)
    elif Chem.MolFromSmiles(row['Smiles']).HasSubstructMatch(Chem.MolFromSmarts("*=CCl")):
        sp2_C.append(row_id)
    else:
        sp3_C.append(row_id)
sp2_C_df = Cl_df.loc[sp2_C]
sp3_C_df = Cl_df.loc[sp3_C]
plot_distribution(ax, sp2_C_df['BDE_G'].to_numpy(), 'top', color_line=color_main, color_fill=color_fill, write_axis = True, frequence=False, bins=15)
plot_distribution(ax, sp3_C_df['BDE_G'].to_numpy(), 'top', color_line=color_2_main, color_fill=color_2_fill, write_axis = True, frequence=False, bins=15)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel("$ BDFE_{C-Cl}$ (kcal/mol)", fontsize=20)
plt.ylabel("Count", fontsize=20)
plt.show()
```

```python
from DFTStructureGenerator.draw import plot_scatter_fit, plot_distribution, plot_panel
from matplotlib import pyplot as plt
fig = plt.figure(figsize=(5,3), dpi=300)
ax = fig.add_subplot()

color_main = "#1f77b4"
color_fill = "#d0e6fa"
color_2_main = "#c04f15"
color_2_fill = "#f2aa84"
color_3_main = "#78206e"
color_3_fill = "#d86ecc"
B_N_df = pd.read_csv(CSV_DIR / "reactants_B_N.csv")
amine_df = B_N_df.loc[np.isin(B_N_df['N_Index'], np.append(Anime_LB, Ar_LB))]
P_df = B_N_df.loc[np.isin(B_N_df['N_Index'], Phosphine_LB)]
NHC_df = B_N_df.loc[np.isin(B_N_df['N_Index'], NHC_LB)]
plot_distribution(ax, amine_df['BDE_G'].to_numpy(), 'top', color_line=color_main, color_fill=color_fill, write_axis = True, frequence=False, bins=15)
plot_distribution(ax, NHC_df['BDE_G'].to_numpy(), 'top', color_line=color_3_main, color_fill=color_3_fill, write_axis = True, frequence=False, bins=15)
plot_distribution(ax, P_df['BDE_G'].to_numpy(), 'top', color_line=color_2_main, color_fill=color_2_fill, write_axis = True, frequence=False, bins=15)

plt.xticks(np.arange(40, 130, 20), fontsize=20)
plt.yticks(fontsize=20)
plt.xlim(40, 125)
plt.xlabel("$ BDFE_{B-Cl}$ (kcal/mol)", fontsize=20)
plt.ylabel("Count", fontsize=20)
plt.show()
```

```python
from DFTStructureGenerator.draw import plot_scatter_fit, plot_distribution, plot_panel
from matplotlib import pyplot as plt
fig = plt.figure(figsize=(5,2), dpi=300)
ax = fig.add_subplot()

color_main = "#1f77b4"
color_fill = "#d0e6fa"
color_2_main = "#c04f15"
color_2_fill = "#f2aa84"
color_3_main = "#78206e"
color_3_fill = "#d86ecc"
B_N_df = pd.read_csv(CSV_DIR / "reactants_B_N.csv")
BH2_smiles, BHR_smiles, BR2_smiles = [], [], []
for B_smiles in B_N_df['B_smiles'].unique():
    B_atom = [atom for atom in Chem.MolFromSmiles(B_smiles).GetAtoms() if atom.GetSymbol() == "B"][0]
    H_neighbors = B_atom.GetTotalNumHs()
    if H_neighbors == 3:
        BH2_smiles.append(B_smiles)
    elif H_neighbors == 2:
        BHR_smiles.append(B_smiles)
    elif H_neighbors == 1:
        BR2_smiles.append(B_smiles)

BH2_df = B_N_df.loc[np.isin(B_N_df['B_smiles'], BH2_smiles)]
BHR_df = B_N_df.loc[np.isin(B_N_df['B_smiles'], BHR_smiles)]
BR2_df = B_N_df.loc[np.isin(B_N_df['B_smiles'], BR2_smiles)]


plot_distribution(ax, BR2_df['BDE_G'].to_numpy(), 'top', color_line=color_3_main, color_fill=color_3_fill, write_axis = True, frequence=1, bins=15)
plot_distribution(ax, BHR_df['BDE_G'].to_numpy(), 'top', color_line=color_2_main, color_fill=color_2_fill, write_axis = True, frequence=1, bins=15)
plot_distribution(ax, BH2_df['BDE_G'].to_numpy(), 'top', color_line=color_main, color_fill=color_fill, write_axis = True, frequence=1, bins=15)

plt.xticks(np.arange(40, 130, 20), fontsize=20)
plt.yticks(fontsize=20)
plt.xlim(40, 125)
plt.xlabel("$ BDE_{B-Cl}$ (kcal/mol)", fontsize=20)
plt.ylabel("Freq.", fontsize=20)
plt.show()
```
